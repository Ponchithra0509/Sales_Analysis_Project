```python
payment_sales = df.groupby("Payment_Method")["Total_Sales"].sum()

payment_sales
```

```python
top_products = df.groupby("Product_Name")["Total_Sales"].sum().sort_values(ascending=False).head(10)

top_products
```

## Data Visualization

```python
monthly_sales.plot(kind="line", marker="o", figsize=(8,5))

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.savefig("../charts/Monthly_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
category_sales.plot(kind="bar", figsize=(8,5), color="skyblue")

plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.savefig("../charts/Category_wise_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_products.plot(kind="barh", figsize=(8,5), color="darkgreen")

plt.title("Top 10 Products")
plt.xlabel("Total Sales")
plt.ylabel("Product Line")

plt.savefig("../charts/Top_10_Products.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
category_sales.plot(
    kind="pie",
    autopct="%1.1f%%",
    explode=(0,0,0,1), 
      shadow=True,          
    figsize=(6,6)
)

plt.title("Category-wise Sales Percentage")
plt.ylabel("")


plt.savefig("../charts/Category_wise_Sales_Percentage.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Total_Sales"], bins=20)

plt.title("Distribution of Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")

plt.savefig("../charts/Distribution_Of_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sns.countplot(data=df, x="Payment_Method")

plt.title("Payment Method Count")

plt.savefig("../charts/Payment_Method_Count.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["Total_Sales"])

plt.title("Box Plot of Total Sales")

plt.savefig("../charts/Box_Plot_of_Total_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Business Insights

## Key Findings

1. The **Electronics** category generated the highest sales with a total revenue of **21,112,771**, making it the best-performing category.

2. **Pune** recorded the highest regional sales with a total revenue of **7,464,811**, followed by Hyderabad and Bangalore.

3. Among all payment methods, **Cash on Delivery (COD)** was the most preferred, generating **12,248,121** in sales.

4. **Laptop** was the highest-selling product, contributing **12,260,253** in total sales.

5. Tablets and Watches were the second and third highest-selling products, showing strong customer demand.

6. The sales analysis helped identify top-performing categories, regions, payment methods, and products, enabling better business decision-making.

7. Data visualization using bar charts, line charts, pie charts, histograms, count plots, and box plots provided a clear understanding of sales performance and customer purchasing patterns.
